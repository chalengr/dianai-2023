```python
import torch
import numpy as np
import pandas as pd
import os
import struct
from torch import nn
```

```python
def read_idx3_ubyte(idx3_ubyte_file):
    with open(idx3_ubyte_file, 'rb') as f:
        magic, num_images, rows, cols = struct.unpack('>IIII', f.read(16))
        images = np.fromfile(f, dtype=np.uint8).reshape(num_images, rows, cols)
    return images

def read_idx1_ubyte(idx1_ubyte_file):
    with open(idx1_ubyte_file, 'rb') as f:
        magic, num_labels = struct.unpack('>II', f.read(8))
        labels = np.fromfile(f, dtype=np.uint8)  
    return labels

```

```python
images = read_idx3_ubyte('train-images.idx3-ubyte')
labels =read_idx1_ubyte('train-labels.idx1-ubyte')
images2 = read_idx3_ubyte('t10k-images.idx3-ubyte')
labels2 =read_idx1_ubyte('t10k-labels.idx1-ubyte')
```
